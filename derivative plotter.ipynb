# Derivative plotter

We can use the `sympy` module to create functions, find derivatives, and plot them. (The plotting isn't great, but it'll do for now.)

### Setup

```python
# import everything from sympy - bad practice (why?)
from sympy import *
# I want to be able to do algebra on x and y
x, y = symbols('x y')
# Display algebra nicely
init_printing()
# Display graphs in the notebook
%matplotlib inline
```

### Define your function

```python
# Define my Cartesian equation
y = x**3+5*x**2+2*x-1
y
```

### Differentiate it

```python
# Differentiate it
dy_by_dx = diff(y)
dy_by_dx
```

### Plot them both

```python
# Plot them both
# Over this domain and range (change this to zoom in)
x_domain = (-5,2)
y_range = (-10,20)
# Plot y= but don't show it yet
p0 = plot(y, xlim=x_domain, ylim=y_range, line_color='blue', show=False)
# Plot the derivative
p1 = plot(dy_by_dx, xlim=x_domain, ylim=y_range, line_color=['red','white'], show=False)
# Put them on the same axes
p0.extend(p1)
# Now show it
p0.show()
```

# The second derivative

If we differentiate *again*, we get the *second derivative*, written $\dfrac{\text{d}^2y}{\text{d}x^2}$

But why would we want to?

```python
d2y_by_dx2 = diff(dy_by_dx)
d2y_by_dx2
```

Plot that too

```python
# Plot them both
# Over this domain and range (change this to zoom in)
x_domain = (-5,2)
y_range = (-20,20)
# Plot y= but don't show it yet
p0 = plot(y, xlim=x_domain, ylim=y_range, line_color='blue', show=False)
# Plot the derivative
p1 = plot(dy_by_dx, xlim=x_domain, ylim=y_range, line_color=['red','white'], show=False)
# Plot the second derivative
p2 = plot(d2y_by_dx2, xlim=x_domain, ylim=y_range, line_color=['green','white'], show=False)
# Put them on the same axes
p0.extend(p1)
p0.extend(p2)
# Now show it
p0.show()
```
